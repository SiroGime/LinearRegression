```python
data = pd.read_csv("/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv")
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
data.plot(kind="scatter", x="YearsExperience", y="Salary")
```

```python
import statsmodels.formula.api as smf
```

```python
lm = smf.ols(formula="Salary~YearsExperience", data=data).fit()
```

```python
lm.params
```

```python
lm.pvalues
```

```python
lm.summary()
```

```python
salary_predict = lm.params.Intercept + lm.params.YearsExperience * data["YearsExperience"]
salary_predict
```

```python
salary_predict = lm.predict(pd.DataFrame(data["YearsExperience"]))
salary_predict
```

```python
SSD = sum((data["Salary"]-salary_predict)**2)
RSE = np.sqrt(SSD/(len(data)-2))
RSE
```

```python
error = RSE/data["Salary"].mean()
error
```

```python
plt.plot(data["YearsExperience"],data["Salary"], "ro")
plt.plot(data["YearsExperience"], salary_predict, color="blue")
```

```python
a = np.random.randn(len(data))
check = (a<0.8)
training = data[check]
testing = data[~check]
len(training), len(testing)
```

```python
lm = smf.ols(formula="Salary~YearsExperience", data=training).fit()
```

```python
lm.summary()
```

```python
salary_predict = lm.predict(pd.DataFrame(testing["YearsExperience"]))
salary_predict
```

```python
SSD = sum((testing["Salary"]-salary_predict)**2)
RSE = np.sqrt(SSD/(len(testing)-2))
error = RSE/testing["Salary"].mean()
error
```
